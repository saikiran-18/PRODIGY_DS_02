```python

import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('/content/Titanic-Dataset.csv')
# Display the first few rows of the dataset
print(df.head())

# Get information about the dataset
print(df.info())

# Summary statistics for numerical columns
print(df.describe())
# Check for missing values
print(df.isnull().sum())

# Handle missing values:
# For simplicity, we can drop rows with missing values
# Alternatively, you might fill missing values with a strategy like mean, median, or mode

# Dropping rows with any missing values
df_cleaned = df.dropna()

# Alternatively, fill missing values for specific columns
# For example, filling missing 'Age' with the median age
# df['Age'].fillna(df['Age'].median(), inplace=True)
# Convert 'Sex' to numeric (0 for male, 1 for female)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Convert 'Embarked' to numeric (using one-hot encoding)
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
# Descriptive statistics for numerical columns
print(df.describe())

# Frequency count for categorical features
print(df['Pclass'].value_counts())
print(df['Sex'].value_counts())
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics for seaborn
sns.set(style="whitegrid")

# Plot distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Survival rate by class
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate by sex
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex (0=Male, 1=Female)')
plt.ylabel('Survival Rate')
plt.show()

# Age vs. Fare scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='coolwarm')
plt.title('Age vs. Fare (colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

```
